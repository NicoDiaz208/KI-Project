# TRAINING

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
import datetime as dt
```

## Date Convertion

```python
data['Year'] = pd.to_datetime({'year': data['Year'],
                   'month': np.full(len(data['Year']),1),
                   'day': np.full(len(data['Year']),1)})

data['Year'] = data['Year'].map(dt.datetime.toordinal)
data['Year']
```

## Implement Model and Train for Primary energy consumption (TWh)

```python
y = data['Primary energy consumption (TWh)']
X = data[['Year']]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=21)
dt_model = LinearRegression()
dt_model.fit(X_train, y_train)
score = dt_model.score(X_train, y_train)

print(score*100)
```

## Implement Model and Train for Electricity Generation (TWh)

```python
y = data['Electricity Generation (TWh)']
X = data[['Year']]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=21)
dt2_model = LinearRegression()
dt2_model.fit(X_train, y_train)
score = dt2_model.score(X_train, y_train)

print(score*100)
```

## Predict For 2021...

```python
d = dt.datetime.strptime('2021-01-01', '%Y-%m-%d').date()

print('Prediction of Consumption and Generation for 2021')
print('Consumption: ',dt_model.predict([[d.toordinal()]]))
print('Generation: ',dt2_model.predict([[d.toordinal()]]))
```
